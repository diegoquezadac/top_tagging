```python
import h5py
import numpy as np
```

```python
input_file = './data/test-public.h5'
output_file = './data/test-public-small.h5'
sample_size = 10000
```

```python
with h5py.File(input_file, 'r') as f_in:
    total_size = f_in['labels'].shape[0]
    
    random_indices = np.random.choice(total_size, size=sample_size, replace=False)
    sorted_idx = np.sort(random_indices)
    
    with h5py.File(output_file, 'w') as f_out:
        for key, value in f_in.attrs.items():
            f_out.attrs[key] = value
        
        for key in f_in.keys():
            data = f_in[key]
            if data.shape[0] == total_size:
                sampled = data[sorted_idx]
                inverse = np.argsort(random_indices)
                sampled = sampled[inverse]
            else:
                sampled = data[:]
            
            f_out.create_dataset(
                key,
                data=sampled,
                compression='gzip',
                compression_opts=4
            )
```
