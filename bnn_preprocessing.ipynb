```python
import time
from src.bnn.dataset import TabularDataset
from src.bnn.preprocess import preprocess
from torch.utils.data import DataLoader
```

```python
preprocess("./data/test-public-small.h5", "./data/train-preprocessed.h5", batch_size=10_000, use_train_weights=True)
```

```python
dataset = TabularDataset("./data/test-public-small-preprocessed.h5", use_train_weights=True)
```

```python
train_loader = DataLoader(
        dataset,
        batch_size=256,
        shuffle=True,
    )

start = time.time()
batch = next(iter(train_loader))
end = time.time()

if isinstance(batch, (list, tuple)):
    x, y, z = batch
    print(f"x.shape: {x.shape}, y.shape: {y.shape}, x.shape: {z.shape}")
else:
    print(f"batch.shape: {batch.shape}")

print(f"Tiempo para cargar un batch: {end - start:.4f} segundos")
```
